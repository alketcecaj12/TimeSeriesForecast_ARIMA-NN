```python
def split_sequence(data, n_input, n_output):
    X, Y = list(), list()
    
    for i in range(len(data)):
        end_ix = n_input + i
        out_end_ix = end_ix + n_output
        
        if out_end_ix > len(data):
            break
            
        seq_x, seq_y = data[i:end_ix], data[end_ix:]
        
        X.append(seq_x)
        Y.append(seq_y)
        
    return X,Y           
```

```python
X,Y = split_sequence(serie, 3, 3)
```
